# Evaluating accuracy

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D, MaxPooling2D, Flatten
(train_img,train_lbl),(test_img,test_lbl)= mnist.load_data()
train_img= train_img.reshape(-1,28,28,1).astype('float32')/255.0
test_img= test_img.reshape(-1,28,28,1).astype('float32')/255.0
model= Sequential([Conv2D(32,kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
                   MaxPooling2D(pool_size=(2, 2)),
                   Flatten(),
                    Dense(128, activation='relu'),
                    Dense(10, activation='softmax')])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(train_img,train_lbl,epochs=5)
loss,accuracy= model.evaluate(test_img,test_lbl)
print("accuracy is ",accuracy)
```

## getting the images in graph format

```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D, MaxPooling2D, Flatten
import matplotlib.pyplot as plt
%matplotlib inline
(train_img,train_lbl),(test_img,test_lbl)= mnist.load_data()
train_img= train_img.reshape(-1,28,28,1).astype('float32')/255.0
test_img= test_img.reshape(-1,28,28,1).astype('float32')/255.0
model= Sequential([Conv2D(32,kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
                   MaxPooling2D(pool_size=(2, 2)),
                   Flatten(),
                    Dense(128, activation='relu'),
                    Dense(10, activation='softmax')])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
model.fit(train_img,train_lbl,epochs=3)
loss,accuracy= model.evaluate(test_img,test_lbl)
print("accuracy is ",accuracy)
print("loss is ", loss)
pred= model.predict([test_img])
for i in range(4):
    print("Enter the index number:")
    ind = int(input())
    print(f"\nThe number at index {ind} is: ")
    print(np.argmax(pred[ind]))
    plt.imshow(test_img[ind],cmap="plasma")
    plt.show()
```

# Using cifar10 dataset for classification

```python
import numpy as np
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D, MaxPooling2D, Flatten
import matplotlib.pyplot as plt
%matplotlib inline
(train_img,train_lbl),(test_img,test_lbl)=cifar10.load_data()
train_img= train_img.reshape(-1,32,32,3).astype('float32')/255.0
test_img= test_img.reshape(-1,32,32,3).astype('float32')/255.0
train_lbl= to_categorical(train_lbl,num_classes=10)
test_lbl= to_categorical(test_lbl,num_classes=10)
model= Sequential([Conv2D(filters=16,kernel_size=(3,3), activation='relu', input_shape=(32,32,3)),
                   MaxPooling2D(pool_size=(2, 2)),
                   Flatten(),
                   Dense(128, activation='relu'),
                   Dense(10, activation='softmax')])
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
print(train_img.shape)
print(train_lbl.shape)
model.fit(train_img,train_lbl,epochs=3)
loss,accuracy= model.evaluate(test_img,test_lbl)
print("accuracy is ",accuracy)
print("loss is ", loss)
pred= model.predict([test_img])
for i in range(4):
    print("Enter the index number:")
    ind = int(input())
    print(f"\nThe number at index {ind} is: ")
    print(np.argmax(pred[ind]))
    plt.imshow(test_img[ind],cmap="plasma")
    plt.show()
```
